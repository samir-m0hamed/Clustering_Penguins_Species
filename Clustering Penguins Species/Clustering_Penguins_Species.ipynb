## Data Preprocessing

## Data Cleaning

```python
data = data[ (data['sex'] != '.') & ( (data['flipper_length_mm'] > 0) & (data['flipper_length_mm'] < 1000) ) ]
```

### Getting The Features

```python
x = data.copy()

x.head()
```

### Encoding The Categorical Data



```python
mapSex = {'FEMALE' : 0 , 'MALE' : 1}

x['sex'] = x['sex'].map(mapSex)
```

```python
x.head()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# 'sex' has already been numerically encoded and should not be scaled further.
columnsToScale = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm' ,'body_mass_g' ]
x[columnsToScale] = scaler.fit_transform(x[columnsToScale])
x.head()
```

## Model Training

### Elbow Method , silhouette_scores

       Elbow Method , It's the method which we will use to get the optimal number for k -> the optimal clusters' number


       silhouette_scores , It's also a method we can use to get the optimal number for k -> the optimal clusters' number

      we caculate silhouette_scores s for each observation in the data by 

      s = ( b - a ) / max ( a, b )

      a -> average eculidean distance between this observation and the other observations in the same cluster

           Minimum a means this observation is close to the points in it's cluster which is good 

      b -> the closest distance between this observation and the the closet other cluster , except the cluster it belongs to

           Maximum b means this observation is far to the other clusters which is good 


      After calculation s for each observation we then get the average for all s
      
      s has values between 1 , -1  

      * s close to 1  -> Good clustering 

      * s close to 0  -> the clusters are overlapped

      * s less that 0 -> not valid 

      * s = [ 0.7 :   1  ] -> very Good 

      * s = [ 0.5 :  0:7 ] -> Good 

      * s = [ 0.25 : 0:5 ] -> average 

      * s = [ 0  :  0:25 ] -> weak 

      * s = [ 0   :  -1  ] -> Not valid 
      

```python
wcss = [] # Within-Cluster Sum of Square

for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10) # n_init=10 by default, k-means++ to avoid random initialization trap
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)   # kmeans.inertia_ ->  wcss value

fig = go.Figure(data=go.Scatter(x=list(range(1, 15)), y=wcss, mode='lines+markers'))
fig.update_layout(title='Elbow Method For Optimal K', xaxis_title='Number of Clusters (K)', yaxis_title='WCSS (Inertia)')
fig.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = {}

# Start k from 2 as silhouette score is not defined for a single cluster
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    clusters = kmeans.fit_predict(x)
    score = silhouette_score( x , clusters )
    silhouette_scores[k] = np.round ( score , 2 )

print(silhouette_scores)
```

### Traing the K-means Model with the optimal K

```python
kmean = KMeans ( n_clusters = 6 , n_init = 10 , init = "k-means++" , random_state = 42  ) # elbow is the main metric

kmean.fit(x)
```

```python
print ( kmean.cluster_centers_ , '\n' )  # centroids position
print ( kmean.labels_ , '\n' )           # clusters that each observation belongs to
print ( kmean.inertia_ )                 # wcss
```

## Visualizing the results

```python
clusters = kmean.predict(x)
data['cluster'] = clusters

data.head(10)
```

```python
fig = px.scatter(data, x='culmen_length_mm', y='culmen_depth_mm', color='cluster',
                 title='Penguin Clusters by Culmen Length and Depth',
                 hover_data=['sex'])
fig.show()
```
